```python
import itertools
from roster import student_roster

# Import modules above this line
class ClassroomOrganizer:
  def __init__(self):
    self.sorted_names = self._sort_alphabetically(student_roster)
  
  def _sort_alphabetically(self,students):
    names = []
    for student_info in students:
      name = student_info['name']
      names.append(name)
    return sorted(names)

  def roll_call(self):
    roll_call_iter = iter(self.sorted_names)
    for student in student_roster:
      next_in_roll_call =  roll_call_iter.__next__()
      return print(list(next_in_roll_call))

  def class_seating(self):
    student_pairs = list(itertools.combinations(self.sorted_names, 2))
    table_combinations = itertools.combinations(student_pairs, 5)
    for table in table_combinations:
     return print(list(table))


  def get_students_with_subject(self, subject):
    selected_students = []
    for student in student_roster:
      if student['favorite_subject'] == subject:
        selected_students.append((student['name'], subject))
    return selected_students

  
```
